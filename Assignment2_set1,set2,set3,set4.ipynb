**2.The current age (in years) of 400 clerical employees at an insurance claims processing center is normally distributed with mean m = 38 and Standard deviation s =6. For each statement below, please specify True/False. If false, briefly explain why.**

A. More employees at the processing center are older than 44 than between 38 and 44.

B. A training program for employees under the age of 30 at the center would be expected to attract about 36 employees.


```python
#  A. More employees at the processing center are older than 44 than between 38 and 44.
Mean_2=38
STD_2=6
#Z for above 44:
Z_44=(44-Mean_2)/STD_2  #Z Score=(Value-Mean)/(STD)
z_44p=stats.norm.cdf(Z_44)
print((1-z_44p)*400)
#People above 44 age:
print('People above the age of 44:',round((1-z_44p)*400,0))

#people between 38 and 44 age:
Z_38=(38-Mean_2)/STD_2    #Z Score=(Value-Mean)/(STD)
z_38p=stats.norm.cdf(Z_38)
Z_44_z_38=round((z_44p-z_38p)*400,0)
print('People between the age of 38 and 44 :',Z_44_z_38)
```

```python
#B. A training program for employees under the age of 30 at the center
# would be expected to attract about 36 employees.

z_30=(30-Mean_2)/STD_2     #Z Score=(Value-Mean)/(STD)

z_30p=stats.norm.cdf(z_30)
#employees under the age of 30 at the center
print(round(z_30p*400,0))
print("True statement is B")
```

True statement B
